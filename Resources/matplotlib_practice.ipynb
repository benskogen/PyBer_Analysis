```python
#Import Dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x-axis = list of strings
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# y-axis - float values of accumulated USD fares
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# create the plot
plt.scatter(x_axis, y_axis)
plt.show()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
plt.show
```

```python
# Create the plot and add a label for the legend. #add markers and colors
plt.plot(x_axis, y_axis, marker="^", color="green", linewidth=2, label='Boston'), plt.grid()
# Create labels for the x and y axes.
plt.xlabel("date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("Pyber Fare by Month")
# Add the legend.
plt.legend(loc="lower center")
plt.figure(figsize=(20, 10))
plt.show()
```

```python
##Create a Vertical Bar Chart
plt.bar(x_axis, y_axis)
plt.show()
```

```python
#calling out the bar chary and customizing
plt.barh(x_axis, y_axis, color="green", label='Boston')
plt.xlabel("date"), plt.ylabel("Fare Money")
plt.title("PyBer fare by month")
plt.legend()
plt.show()
```

```python
#invert y-axis

plt.barh(x_axis, y_axis, color="#b22222", label='Boston')
plt.xlabel("date"), plt.ylabel("Fare Money")
plt.title("PyBer fare by month")
plt.legend()
plt.gca().invert_yaxis()
plt.show
```

```python
#scatter plot w/ with size parameter
plt.scatter(x_axis, y_axis, color="red", s=y_axis)
plt.xlabel("date")
plt.ylabel("Fare$")
plt.title("PyBer fare by month")
plt.show()
```

```python
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
plt.show()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger, color="rebeccapurple")
plt.show()
```

```python
#create color list indexed to month list
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.3, 0, 0, 0, 0, 0,)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, explode=explode_values, colors=colors, labels=x_axis, autopct="%.1f%%")
plt.show()
```
